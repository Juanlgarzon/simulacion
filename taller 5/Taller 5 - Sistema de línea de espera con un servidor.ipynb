# Modelo de colas G/G/1

```python
import random
import math

def aleatorio():
    return random.random()

def expcial(llamda):
    aleatrio = aleatorio()
    value = - (math.log(1 - aleatrio) / float(llamda))
    return value

def normal():
    aleatrio = aleatorio()
    square_root = math.sqrt(-2 * math.log1p(aleatrio - 1))
    sin = math.sin(2 * math.pi * aleatrio)
    value = square_root * sin
    if(value < 0):
        value = 0
    return value
def llega(tact):
    global t
    global nclientes
    global nlleg
    global llegadas
    global llamda
    global T
    global tsuc
    t = tact
    nclientes += 1
    nlleg += 1
    llegadas.append(t)
    X = expcial(llamda)
    if(t + X < T):
        tsuc['tllegadas'] = t + X
    if(nclientes == 1):
        Y = normal()
        tsuc['tsalidas'] = t + Y
        tservicios.append(Y)
        

def servidor(tact):
    global t
    global nclientes
    global nsal
    global salidas
    global tsuc
    t = tact
    nclientes -= 1
    nsal += 1
    salidas.append(t)
    if(nclientes > 0):
        Y = normal()
        tsuc['tsalidas'] = t + Y
        tservicios.append(Y)
t = 0 
nclientes = 0
nlleg = 0
nsal = 0
T = 10
tp = 0
tact = 0

t_med_sistema = 0
t_med_espera = 0

salidas = []
llegadas = []
tservicios = []



M = 9999999999

tsuc = {
    'tllegadas': M,
    'tsalidas': M
}

llamda = 4
tact = expcial(llamda)

if tact > T:
    tp = 0
    t_med_sistema = 0
    t_med_espera = 0
else:
    llega(tact)
    
    while(tsuc['tllegadas'] != M or tsuc['tsalidas'] != M):
        if(tsuc['tllegadas'] < tsuc['tsalidas']):
            X = tsuc['tllegadas']
            tsuc['tllegadas'] = M
            llega(X)
        if(tsuc['tsalidas'] < tsuc['tllegadas']):
            X = tsuc['tsalidas']
            tsuc['tsalidas'] = M
            servidor(X)
    
    tp = max(0, t - T)
    acumulo1 = 0
    acumulo2 = 0
    ind = 0
    while ind < nlleg:
        acumulo1 += salidas[ind] - llegadas[ind]
        acumulo2 += salidas[ind] - llegadas[ind] - tservicios[ind]
        ind += 1
    
    t_med_sistema = acumulo1 / nlleg
    t_med_espera = acumulo2 / nlleg


print("Tiempo de simulación: ",t)
print("Tiempo medio en el sistema:",t_med_sistema)
print("Tiempo medio en la cola: ",t_med_espera)
```

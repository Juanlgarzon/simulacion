```python
## Metodos de Montecarlo con limites
import random
import math
import sympy as sp



def mt1(fu,n):
    #fu=input("Introduce la funcion a evaluar: ");
    suma=0;
    i=0;
    while i < n:
    
        u=random.random();
        #x=u;
        #y=eval(fu)
        y=sp.sympify(fu).subs(x,u);
        #y=math.exp(-u*u);
        suma+=y;
        i += 1;
    #print(suma); 
    return(suma/n); 




def mt2(fu,n,a,b):
    #fu=input("Introduce la funcion a evaluar: ");
    suma=0;
    i=0;
    dist=b-a;
    while i < n:
    
        
        ra=(a+(dist*random.random()));
        #x=ra;
        #y=dist*(eval(fu));
        y=sp.sympify(fu).subs(x,ra);
        suma+=y;
        i += 1;
    #print(suma);
    return(dist*suma/n); 

def mt3(fu,n):
    #fu=input("Introduce la funcion a evaluar: ");
    suma=0;
    i=0;
    while i < n:
        u=random.random();
        ra=(1/u)-1;
        #x=ra
        #y=(1/(ra*ra))*(eval(fu));
        y=(1/(u*u))*sp.sympify(fu).subs(x,ra);
        suma+=y;
        i += 1;
    #print(suma); s
    return(suma/n); 


n=float(input("Introduce el valor de n para el metodo de Montecarlo: "));
x = sp.Symbol('x')
#e=math.exp(1);
fu=input("Introduce la funcion a evaluar en terminos de x: ");
fu=fu.replace("e",str(math.exp(1)));
print("Para infinito= 'inf' ");
a=int(input("Introduce el valor de a: "));

b=(input("Introduce el valor de b: "));


r=float();


if(b=="inf"):

    r=mt3(fu,n);
    #print(r);
else:
    b=int(b);
    
    if(a==0 and b==1):
        
        r=mt1(fu,n);
    else:
        
        r=mt2(fu,n,a,b);
#print(r);        

print("El resultado mediante el metodo de Montecarlo= ",r);

```

```python
## Aproximacion de PI
import random

    
def mids(n):
    #x=float(input("Ingrese el valor de la semilla x0:\n "));
    #x2=float(input("Ingrese el valor de la segunda semilla x0:\n "));
    i=0;
    #print("\nLos puntos generados son: ");
    #print("X   Y")
    d=0;
    
    prueba=[];
    while i < n:
  
        
        u=round(random.random(),4);
        u2=round(random.random(),4);
        
        i=i+1;
        a=(2*u)-1;
        b=(2*u2)-1;
        c=(a**2)+(b**2);
        
        if(c<=1):
            texto="El punto "+str(u)+","+str(u2)+" esta dentro del circulo";
            prueba.append(texto);
            d=d+1;
        
        else:
            texto="El punto "+str(u)+","+str(u2)+" esta fuera del circulo";
            prueba.append(texto);
    
   
    aprox=(4*d)/n;
    print("La aproximacion de pi con ",n," es= ",aprox);
    prueba.append("\nPuntos dentro del circulo: "+str(d));
    prueba.append("Puntos fuera del circulo: "+str(n-d));
    return prueba;





n=int(input("Ingrese la cantidad maxima de puntos aleatorios a generar: "));

letters=(mids(n));

print("Desea ver los puntos aleatorio que estan dentro del circulo y los que no?\nY==1\tN==0\n")
m=float(input());
print("\n")
if(m==1):
    for i in letters:
        print(i)
    

```
